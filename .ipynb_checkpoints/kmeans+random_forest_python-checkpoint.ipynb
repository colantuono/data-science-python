```python
func = pd.DataFrame(df.groupby('ID').agg('mean'))
func.drop(['Reason for absence', 'Month of absence', 'Day of the week', 'Seasons', 'Disciplinary failure', 'Hit target'], axis = 1, inplace=True)
# func.drop('Absenteeism time in hours', axis=1, inplace=True)
func
```

### DATA SPLIT

```python
x = func.drop('Absenteeism time in hours', axis= 1)
y = func['Absenteeism time in hours']

x_train, x_test, y_train, y_test = tts(x, y, test_size=0.5, random_state = 1)
```

### Evaluating correct number of clusters

```python
min_clusters = 2
max_clusters = 15

inertias = np.zeros(shape = (max_clusters - min_clusters + 1))
```

```python
for i in range(min_clusters, max_clusters + 1):
    km = KMeans(n_clusters = i, random_state = 1)
    km.fit(x_train)
    inertias[i - min_clusters] = km.inertia_

plt.figure(figsize = (10, 6))
plt.scatter(range(min_clusters, max_clusters + 1), inertias, color = "b")
plt.title("Avalia√ßao do numero de clusters")
plt.xlabel("# de clusters")
plt.ylabel("Soma da distancia quadrada")
# plt.annotate("# de clusters escolhido", xy = (6, 1.25e7), xytext = (7.5, 2e7),
#             arrowprops = dict(facecolor = "red", shrink = 0.05),
#             );
```

```python
kmodel = KMeans(n_clusters = 4, random_state=1)

clusters = kmodel.fit(x_train)
train_clusters = pd.DataFrame(clusters.labels_,columns = ["Cluster"])
train_clusters["Label"] = y_train.values

clusters.labels_
```

Applying KMeans to new data

```python
fitted_clusters = kmodel.predict(x_test)
test_clusters = pd.DataFrame(fitted_clusters,columns = ["Cluster"])
test_clusters["Label"] = y_test.values
test_clusters
```

```python
clusters_df = pd.DataFrame(fitted_clusters,columns = ["Cluster"], )
print(clusters_df.shape);print(x_test.shape)

test_clusters_array = np.array(test_clusters['Cluster'].values.tolist())

teste = x_test
teste['Cluster'] = test_clusters_array
teste.head()
```
